```python
import pandas as pd
import os
from collections import defaultdict, Counter
from datetime import datetime, timedelta

# label을 덮어 씌울 submission.csv 아무거나
base_df = pd.read_csv('./data/sample_submission.csv')

# ensemble 할 csv들이 있는 폴더 경로
csv_path = './results/soft_ensemble'
```

```python
dirs = os.listdir(csv_path)
dirs = sorted(dirs)
print("@@@@@@@@@@@@@@@@@@@@@")
print(dirs)
print("@@@@@@@@@@@@@@@@@@@@@")

type_dict = defaultdict(list)
polarity_dict = defaultdict(list)
tense_dict = defaultdict(list)
certainty_dict = defaultdict(list)
for i in range(len(base_df)):
    type_dict[i]
    polarity_dict[i]
    tense_dict[i]
    certainty_dict[i]

for path in os.listdir(csv_path):
    if path.endswith('.csv'):
        df = pd.read_csv(os.path.join(csv_path,path))
        df['type'] = df['label'].apply(lambda x : x.split('-')[0])
        df['polarity'] = df['label'].apply(lambda x : x.split('-')[1])
        df['tense'] = df['label'].apply(lambda x : x.split('-')[2])
        df['certainty'] = df['label'].apply(lambda x : x.split('-')[3])

        for i in range(len(df)):
            type_dict[i].append(df['type'][i])
            polarity_dict[i].append(df['polarity'][i])
            tense_dict[i].append(df['tense'][i])
            certainty_dict[i].append(df['certainty'][i])

type_counter = []
polarity_counter = []
tense_counter = []
certainty_counter = []

for key in type_dict.keys():
    type_counter.append(Counter(type_dict[key]).most_common()[0][0])
for key in polarity_dict.keys():
    polarity_counter.append(Counter(polarity_dict[key]).most_common()[0][0])
for key in tense_dict.keys():
    tense_counter.append(Counter(tense_dict[key]).most_common()[0][0])
for key in certainty_dict.keys():
    certainty_counter.append(Counter(certainty_dict[key]).most_common()[0][0])

ensemble_label = []
for i in range(len(base_df)):
    ensemble_label.append(type_counter[i] + "-" + polarity_counter[i] + "-" + tense_counter[i] + "-" + certainty_counter[i])

base_df['label'] = ensemble_label
```

```python
now_date = datetime.now()
diff_hours = timedelta(hours=9)
now_date += diff_hours
print_now = str(now_date.month) + '_' + str(now_date.day) + '_' + str(now_date.hour) + '_' + str(now_date.minute)

if not os.path.exists('./results'):
    os.makedirs('./results')

if not os.path.exists(os.path.join('./results','hard_ensemble')):
    os.makedirs(os.path.join('./results','hard_ensemble'))

base_df.to_csv(os.path.join('./results','hard_ensemble',f'hard_results_{print_now}.csv'), index = False)

print("Inference Finish")
```
